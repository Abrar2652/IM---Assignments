**Assignment 1**

Plot the following equations in the same plot 

1. Logarithmic = O(log n)
2. Linear = O(n)
3. Quadratic = O(n<sup>2</sup>)
4. Cubic = O(n<sup>3</sup>)
5. Polynomial = O(n<sup>k</sup>)

Input Values: n=7, k=4


```python
import numpy as np
import matplotlib.pyplot as plt


# Stylesheets defined in Matplotlib
plt.style.use('bmh')

# Set up runtime comparisons
n = np.linspace(1, 7, 1000)
k = 4
labels = ['Logarithmic', 'Linear', 'Quadratic', 'Cubic', 'Polynomial']
big_o = [np.log(n), n, n**2, n**3, n**k]

# Plot setup
plt.figure(figsize=(12, 10))
plt.ylim(0, 50)

for i in range(len(big_o)):
    plt.plot(n, big_o[i], label=labels[i])

plt.legend(loc=0)
plt.ylabel('Relative Runtime')
plt.xlabel('Input Size')
plt.show()
```
